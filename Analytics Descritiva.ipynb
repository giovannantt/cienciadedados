```python
import pandas as pd
import matplotlib.pyplot as plt

###Gráfico da quantiade de alunos Tipo de ingresso X Cor ###


# Leitura do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Contagem da frequência de cada combinação de tipo_ingresso e cor
frequencia = dengo_df.groupby(['TIPO_INGRESSO', 'COR']).size().unstack()

# Criação o gráfico de barras
frequencia.plot(kind='bar', stacked=True)

# Adicionar título e rótulos aos eixos
plt.title('Gráfico de Barras de Tipo de Ingresso por Cor')
plt.xlabel('Tipo de Ingresso')
plt.ylabel('Frequência')

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

###Gráfico da quantidade de alunos diferenciando Período letivo x Truno ###

# Leitura do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Criação  o gráfico de barras
plt.figure(figsize=(10, 6))
sns.countplot(data=dengo_df, x='PERIODO_LETIVO', hue='TURNO')

# Criação do título e rótulos
plt.title('Gráfico de Barras de Período Letivo por Turno')
plt.xlabel('Período Letivo')
plt.ylabel('Contagem')

# Exibição da legenda
plt.legend(title='Turno')

# Ajuste do layout para melhor visualização
plt.tight_layout()

# Exibição do gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

###Gráfico da faixa etária dos alunos por Campus###

# Leitra do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Calculo da idade dos alunos
ano_atual = 2024  # Assumindo que estamos em 2024
dengo_df['IDADE'] = ano_atual - dengo_df['ANO_NASCIMENTO']

# Definição das faixas etárias
faixas_etarias = [0, 20, 30, 40, 50, 60, 100]  # Exemplo de faixas etárias
rotulos_faixas_etarias = ['15-20', '21-30', '31-40', '41-50', '51-60', '61+']

# Criação de uma nova coluna 'FAIXA_ETARIA' com base nas faixas etárias
dengo_df['FAIXA_ETARIA'] = pd.cut(dengo_df['IDADE'], bins=faixas_etarias, labels=rotulos_faixas_etarias)

# Criação do gráfico de barras por campus
plt.figure(figsize=(12, 6))
sns.countplot(data=dengo_df, x='FAIXA_ETARIA', hue='CAMPUS')

# Criação do título e rótulos
plt.title('Gráfico de Barras de Faixa Etária por Campus')
plt.xlabel('Faixa Etária')
plt.ylabel('Contagem')

# Ajuste do layout para melhor visualização
plt.tight_layout()

# Exibição do gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

###Gráfico do Top10 cidades com mais matriculados dos campus ###

# Leitura do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Calculo das estatísticas de matriculados por cidade
estatisticas_por_cidade = dengo_df['CIDADE_NAT'].value_counts()

# Ordenação das cidades por número de matriculados
estatisticas_por_cidade = estatisticas_por_cidade.sort_values(ascending=False)

# Criação de ranking Top 10 cidades com mais matriculados
top_10_cidades = estatisticas_por_cidade.head(10)

# Criação do gráfico de barras
plt.figure(figsize=(12, 6))
top_10_cidades.plot(kind='bar', color='skyblue')

# Título e rótulos
plt.title('Top 10 Cidades com Mais Matriculados')
plt.xlabel('Cidade')
plt.ylabel('Número de Matriculados')

# Exibição do gráfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

###Gráfico da distribuição de estrangeiros da universidade###

# Leitura do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Filtro de estrangeiros matriculados nos campus
estrangeiros = dengo_df[dengo_df['NACIONALIDADE'] != 'BRA']

# Contagem do número de estrangeiros por país
contagem_por_pais = estrangeiros['NACIONALIDADE'].value_counts()

# Criação do gráfico de pizza
plt.figure(figsize=(10, 6))
contagem_por_pais.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribuição de Estrangeiros Matriculados por País')

# Exibir o gráfico
plt.axis('equal')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

###Gráfico de proporção de estudantes PCD###


# Leitura do DataFrame do arquivo CSV
dengo_df = pd.read_csv("/content/drive/My Drive/dados_mba/base.csv")

# Contagem do número de estudantes com deficiência
estudantes_com_deficiencia = dengo_df[dengo_df['DEFIC'] == 'SIM']['ID_DISCENTE'].count()

# Contagem do número total de estudantes
total_estudantes = dengo_df['ID_DISCENTE'].nunique()

# Calculo da proporção de estudantes com deficiência
proporcao_deficiencia = estudantes_com_deficiencia / total_estudantes

# Criação de um DataFrame para o gráfico de pizza
dados = pd.DataFrame({'Categoria': ['Com Deficiência', 'Sem Deficiência'],
                      'Proporção': [proporcao_deficiencia, 1 - proporcao_deficiencia]})

# Criação do gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(dados['Proporção'], labels=dados['Categoria'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=140)
plt.title('Proporção de Estudantes com Deficiência')

# Exibição do gráfico
plt.axis('equal')
plt.show()
```
